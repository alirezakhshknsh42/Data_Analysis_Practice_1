Online Retail Sales Analysis

```python
# Importing necessary libraries
import numpy as np  # For numerical operations and working with arrays
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For plotting and data visualization
import matplotlib as mpl  # Importing the matplotlib library as mpl
from mpl_toolkits.mplot3d import Axes3D  # Importing the Axes3D class from mpl_toolkits.mplot3d module
%matplotlib inline
```

```python
# The product categories are provided in the problem statement. The following array is used to hold these categories.
product_categories = np.array(['Furniture', 'Office Supplies', 'Technology'])
```

```python
# The following array is used to hold the sales amounts for each product category.
sales_amounts = np.array([4110451.90, 4744557.50, 3787492.52])
```

```python
# Plotting the bar graph
mpl.rcParams['font.family'] = 'Arial'  # Sets the global font family to Arial
plt.figure(figsize=(4,2))  # Optional: Sets the figure size to 4x2 inches
plt.bar(product_categories, sales_amounts, color=['blue', 'green', 'red'])  # Creates a bar graph with specified colors for each bar
plt.title('Sales Amount by Product Category', fontweight = 'bold')  # Sets the title of the graph
plt.xlabel('Product Category')  # Sets the label for the x-axis
plt.ylabel('Sales Amount')  # Sets the label for the y-axis
plt.grid(False)  # Optional: Adds a grid to the graph for better readability
plt.show()  # Displays the graph
```

```python
# modifying the bars in the graph
# changing color of the bars in the bar graph
mpl.rcParams['font.family'] = 'Arial'  # Sets the global font family to Arial
plt.figure(figsize=(4,2))  # Optional: Sets the figure size to 4x2 inches
plt.bar(product_categories, sales_amounts, color=['purple', 'orange', 'pink'])  # Creates a bar graph with specified colors for each bar
plt.title('Sales Amount by Product Category', fontweight = 'bold')  # Sets the title of the graph
plt.xlabel('Product Category')  # Sets the label for the x-axis
plt.ylabel('Sales Amount')  # Sets the label for the y-axis
plt.grid(False)  # Optional: Adds a grid to the graph for better readability
plt.show()  # Displays the graph
```

```python
plt.figure(figsize=(4,2))  # Optional: Sets the figure size to 4x2 inches
plt.barh(product_categories, sales_amounts, color=['purple', 'orange', 'pink'])  # Creates a horizontal bar graph with specified colors for each bar
plt.title('Sales Amount by Product Category', fontweight = 'bold')  # Sets the title of the graph
plt.xlabel('Sales Amount')  # Sets the label for the x-axis
plt.ylabel('Product Category')  # Sets the label for the y-axis
plt.grid(False)  # Optional: Adds a grid to the graph for better readability
plt.show()  # Displays the graph
```

```python
plt.figure(figsize=(4,2))  # Optional: Sets the figure size to 4x2 inches
# Assume sales_amounts is the data you are plotting
sales_amounts = [4110451.90, 4744557.50, 3787492.52]
# Convert sales_amounts to lakhs for better readability
sales_amounts_in_lakhs = [amount / 100000 for amount in sales_amounts]
# Create a new plot
plt.bar(product_categories, sales_amounts_in_lakhs, color=['purple', 'orange', 'pink'])
# Modifying the ticks to show information in (lakhs)
tick_values = np.arange(0, max(sales_amounts_in_lakhs)+10, 10)  # Creates tick values from 0 to the maximum value in sales_amounts_in_lakhs with a step of 10
tick_labels = [f'{value}L' for value in tick_values]  # Creates labels for each tick value appending 'L' to indicate lakhs
# Apply the new tick values and labels to the y-axis
plt.yticks(tick_values, tick_labels)
plt.title('Sales Amount by Product Category', fontweight = 'bold')  # Sets the title of the graph
plt.xlabel('Product Category')  # Sets the label for the x-axis
plt.ylabel('Sales Amount')  # Sets the label for the y-axis
plt.grid(False)  # Optional: Adds a grid to the graph for better readability
# Display the plot
plt.show()
```

```python
# Creating a 3D bar plot
fig = plt.figure(figsize=(10,4)) # Sets the figure size to 10x4 inches
ax = fig.add_subplot(111, projection='3d')  # Creates a 3D axis object
# Defining x, y, and z coordinates for the bars
x = range(len(product_categories))  # x-coordinates of the bars
y = [1] * len(product_categories)  # y-coordinates of the bars
z = [0] * len(product_categories)  # z-coordinates of the bars
# Defining the width, depth, and height of the bars
dx = [0.5] * len(product_categories)  # Width of the bars
dy = [0.5] * len(product_categories)  # Depth of the bars
dz = sales_amounts  # Height of the bars (sales amounts)
# Plotting the 3D bar chart
ax.bar3d(x, y, z, dx, dy, dz, color=['blue', 'green', 'red'])  # Creates a 3D bar chart
# Setting the labels and title
ax.set_xticks(range(len(product_categories)))  # Sets the x-tick positions
ax.set_xticklabels(product_categories)  # Sets the x-tick labels
ax.set_title('Sales Amount by Product Category', fontweight = 'bold')  # Sets the title of the graph
ax.set_ylabel('Product Category')  # Sets the label for the y-axis (depth)
ax.set_zlabel('Sales Amount')  # Sets the label for the z-axis (height)
# Display the 3D bar chart
plt.show()  # Displays the graph
```
